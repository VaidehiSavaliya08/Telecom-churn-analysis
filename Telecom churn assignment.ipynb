```python
# Calculating total recharge amount
telecom['total_data_rech_6'] = telecom.total_rech_data_6 * telecom.av_rech_amt_data_6
telecom['total_data_rech_7'] = telecom.total_rech_data_7 * telecom.av_rech_amt_data_7
```

```python
# apply scaling on the dataset
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train[num_col] = scaler.fit_transform(X_train[num_col])
```

### PCA

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pca = Pipeline([('scaler', StandardScaler()), ('pca', PCA())])
```

```python
pca.fit(X_train)
churn_pca = pca.fit_transform(X_train)
```

```python
# extracting pca model from pipeline
pca = pca.named_steps['pca']

# checking explainded variance of PCA components
print(pd.Series(np.round(pca.explained_variance_ratio_.cumsum(), 4)*100))
```

```python
cumulative_variance = np.round(np.cumsum(pca.explained_variance_ratio_)*100, decimals=4)
```

```python
fig = plt.figure(figsize=[10,6])
plt.vlines(x=15, ymax=1, ymin=0, colors="r", linestyles="--")
plt.hlines(y=0.95, xmax=30, xmin=0, colors="g", linestyles="--")
plt.plot(cumulative_variance)
plt.ylabel("Cumulative variance explained")
plt.show()
```

### PCA and Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
```

```python
# creating pipeline
PCA_VARS = 60
steps = [('scaler', StandardScaler()),
         ("pca", PCA(n_components=PCA_VARS)),
         ("logistic", LogisticRegression(class_weight='balanced'))
        ]
pipeline = Pipeline(steps)
```

```python
# fiting model
pipeline.fit(X_train, y_train)

# checking score on train data
pipeline.score(X_train, y_train)
```

#### Hyperparameter tuning - PCA and Logistic Regression


```python
# checking class imbalance
y_train.value_counts()/y_train.shape
```

```python
# Applying PCA
pca = PCA()
```

```python
# Using logistic regression - the class weight is used to handle class imbalance as it adjusts the cost function
logistic = LogisticRegression(class_weight={0:0.1, 1: 0.9})
```

```python
# creating pipeline
steps = [("scaler", StandardScaler()), 
         ("pca", pca),
         ("logistic", logistic)
        ]
```

```python
# compiling pipeline
pca_logistic = Pipeline(steps)
```

```python
# setting hyperparameter space
params = {'pca__n_components': [60, 80], 'logistic__C': [0.1, 0.5, 1, 2, 3, 4, 5, 10], 'logistic__penalty': ['l1', 'l2']}
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
# creating 5 folds
folds = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 4)
```

```python
# creating gridsearch object
model = GridSearchCV(estimator=pca_logistic, cv=folds, param_grid=params, scoring='roc_auc', n_jobs=-1, verbose=1)
```

```python
# fiting the model
model.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
print("Best AUC: ", model.best_score_)
print("Best hyperparameters: ", model.best_params_)
```

```python
# predicting churn on test data
y_pred = model.predict(X_test)

# creating onfusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# checking sensitivity and specificity
sensitivity, specificity, _ = sensitivity_specificity_support(y_test, y_pred, average='binary')
print("Sensitivity: \t", round(sensitivity, 2), "\n", "Specificity: \t", round(specificity, 2), sep='')

# checking area under curve
y_pred_prob = model.predict_proba(X_test)[:, 1]
print("AUC:    \t", round(roc_auc_score(y_test, y_pred_prob),2))
```
